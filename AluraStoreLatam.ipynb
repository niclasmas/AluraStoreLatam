### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_total_tienda1 = tienda['Precio'].sum()
ingreso_total_tienda2 = tienda2['Precio'].sum()
ingreso_total_tienda3 = tienda3['Precio'].sum()
ingreso_total_tienda4 = tienda4['Precio'].sum()
```

```python
print(ingreso_total_tienda1)
print(ingreso_total_tienda2)
print(ingreso_total_tienda3)
print(ingreso_total_tienda4)
```

```python
import matplotlib.pyplot as plt
ingresos = [tienda['Precio'].sum(), tienda2['Precio'].sum(),
            tienda3['Precio'].sum(), tienda4['Precio'].sum()]
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
```

```python
plt.barh(nombres, ingresos, color='green')
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Suma de Precio ($)')
plt.show()
```

# 2. Ventas por categoría

```python
ventas_cat_tienda1 = tienda['Categoría del Producto'].value_counts()
ventas_cat_tienda2 = tienda2['Categoría del Producto'].value_counts()
ventas_cat_tienda3 = tienda3['Categoría del Producto'].value_counts()
ventas_cat_tienda4 = tienda4['Categoría del Producto'].value_counts()
```

```python
print(" Ventas por Categoría: Tienda 1 ")
print(ventas_cat_tienda1)

print(" Ventas por Categoría: Tienda 2 ")
print(ventas_cat_tienda2)

print(" Ventas por Categoría: Tienda 3 ")
print(ventas_cat_tienda3)

print(" Ventas por Categoría: Tienda 4 ")
print(ventas_cat_tienda4)
```

# 3. Calificación promedio de la tienda


```python
promedio_t1 = tienda['Calificación'].mean()
promedio_t2 = tienda2['Calificación'].mean()
promedio_t3 = tienda3['Calificación'].mean()
promedio_t4 = tienda4['Calificación'].mean()
```

```python
print(f"Promedio Tienda 1: {promedio_t1:.2f}")
print(f"Promedio Tienda 2: {promedio_t2:.2f}")
print(f"Promedio Tienda 3: {promedio_t3:.2f}")
print(f"Promedio Tienda 4: {promedio_t4:.2f}")
```

```python
import matplotlib.pyplot as plt

calificaciones = [tienda['Calificación'].mean(),
                  tienda2['Calificación'].mean(),
                  tienda3['Calificación'].mean(),
                  tienda4['Calificación'].mean()]

nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.plot(nombres, calificaciones, marker='o', color='green')

plt.title('Calificación Promedio por Tienda')
plt.ylabel('Estrellas (Promedio)')

plt.show()

```

# 4. Productos más y menos vendidos

```python
ventas_t1 = tienda['Producto'].value_counts()
print("Tienda 1 - Más vendido:", ventas_t1.index[0], "-", ventas_t1.values[0], "unidades")
print("Tienda 1 - Menos vendido:", ventas_t1.index[-1], "-", ventas_t1.values[-1], "unidades")
print("-" * 30)

ventas_t2 = tienda2['Producto'].value_counts()
print("Tienda 2 - Más vendido:", ventas_t2.index[0], "-", ventas_t2.values[0], "unidades")
print("Tienda 2 - Menos vendido:", ventas_t2.index[-1], "-", ventas_t2.values[-1], "unidades")
print("-" * 30)

ventas_t3 = tienda3['Producto'].value_counts()
print("Tienda 3 - Más vendido:", ventas_t3.index[0], "-", ventas_t3.values[0], "unidades")
print("Tienda 3 - Menos vendido:", ventas_t3.index[-1], "-", ventas_t3.values[-1], "unidades")
print("-" * 30)

ventas_t4 = tienda4['Producto'].value_counts()
print("Tienda 4 - Más vendido:", ventas_t4.index[0], "-", ventas_t4.values[0], "unidades")
print("Tienda 4 - Menos vendido:", ventas_t4.index[-1], "-", ventas_t4.values[-1], "unidades")
```

# 5. Envío promedio por tienda

```python
envio_t1 = tienda['Costo de envío'].mean()
envio_t2 = tienda2['Costo de envío'].mean()
envio_t3 = tienda3['Costo de envío'].mean()
envio_t4 = tienda4['Costo de envío'].mean()
```

```python
print(f"Costo de envío promedio Tienda 1: ${envio_t1:,.2f}")
print(f"Costo de envío promedio Tienda 2: ${envio_t2:,.2f}")
print(f"Costo de envío promedio Tienda 3: ${envio_t3:,.2f}")
print(f"Costo de envío promedio Tienda 4: ${envio_t4:,.2f}")
```

```python
import matplotlib.pyplot as plt

envio_promedios = [tienda['Costo de envío'].mean(),
                   tienda2['Costo de envío'].mean(),
                   tienda3['Costo de envío'].mean(),
                   tienda4['Costo de envío'].mean()]

nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.bar(nombres, envio_promedios, color='green')
plt.title('Costo de Envío Promedio')
plt.show()
```

**1. Introducción**

El propósito de este análisis no es solo revisar cifras, sino entender la salud operativa de Alura Store. Para ayudar al Sr. Juan en su decisión de venta, hemos evaluado la rentabilidad, la conexión con los clientes y la eficiencia de cada ubicación, buscando identificar qué tienda tiene el mejor potencial y cuál representa una carga para el crecimiento de la marca.

**2. Hallazgos Principales**

Tras cruzar los datos de ventas, logística y satisfacción, estos son los puntos clave:

Rendimiento Comercial: Existe una diferencia clara en el dinamismo de las tiendas. Mientras que las Tiendas 1, 2 y 3 mantienen un flujo de ingresos saludable y constante, la Tienda 4 se queda rezagada, mostrando el volumen de ventas más bajo de toda la red.

La Voz del Cliente: No todo es dinero; la opinión de quienes compran es fundamental. El análisis de reseñas muestra que la Tienda 4 tiene el promedio de satisfacción más bajo. Esto sugiere que los clientes en esa ubicación no están teniendo una buena experiencia, lo que daña la reputación general de la marca.

El Desafío Logístico: Un factor crítico es el costo de envío. Hemos detectado que la Tienda 4 enfrenta los costos de envío más altos en promedio. Esto crea un círculo vicioso: el cliente paga más por recibir su producto y, posiblemente debido a ese costo extra, califica peor a la tienda.

**3. Conclusión y Recomendación Estratégica**

Después de evaluar cada factor, mi recomendación profesional para el Sr. Juan es proceder con la venta de la Tienda 4.Ya que la Tienda 4 presenta lo que en análisis de datos llamamos un "rendimiento ineficiente": genera pocos ingresos pero tiene los costos de (envío) más elevados. Mantenerla implica gastar energía y recursos en una ubicación que no solo es menos rentable, sino que también está afectando el promedio de satisfacción de sus clientes.
